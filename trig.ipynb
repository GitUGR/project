```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid()
```

```python
y = np.cos(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid()
```

```python
y = np.tan(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
y = np.sin(x)/2+np.cos(x)**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
